Open a scanned COVID image and the corresponding mask in matplotlib. Check the sizes and properties.
See: https://stackoverflow.com/a/65507714

```python
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# parameters
condition_2plot = "COVID"
num_patient = 10

file_name = condition_2plot + '-' + str(num_patient) + '.png'
# The file name is shared by image and the corresponding mask.

# paths
path_data = os.path.join("../data", condition_2plot)
path_images = os.path.join(path_data, "images")
path_image = os.path.join(path_images, file_name)

# load the image
scan_img = mpimg.imread(path_image)
print(type(scan_img))  # ndarray
print(scan_img.shape)  # (299, 299)
```

```python
plt.imshow(scan_img)
plt.colorbar()
```

```python
# Load the mask
path_masks = os.path.join(path_data, "masks")
path_mask = os.path.join(path_masks, file_name)

# load the mask
mask_img = mpimg.imread(path_mask)
print(type(mask_img))  # ndarray
print(mask_img.shape)  # (256, 256, 3)
```

```python
plt.imshow(mask_img)
```

According to the plot, the mask contains only 2 colors (black and white). Do the histograms of each color channel show just 2 values as well?

```python
plt.hist(mask_img[:, :, 0].ravel(), color='red')
```

```python
plt.hist(mask_img[:, :, 1].ravel(), color='green')
```

```python
plt.hist(mask_img[:, :, 2].ravel(), color='blue')
```

Indeed, each histogram shows only 2 values and the numbers are identical.
